## CONEXION BASE DE DATOS

```python
#  pandas actualizado correctamente
!pip install --upgrade pandas

# importar pandas y sqlite3
import pandas as pd
import sqlite3 as sql

# Crear una conexión a la base de datos SQLite
conn = sql.connect('/content/drive/MyDrive/analitica3win/proyecto_recursoshumanos/base.db')

# Consultar las primeras filas de cada tabla
print("\nDatos de la tabla 'retiros':")
df_retiros = pd.read_sql("SELECT * FROM retiros", conn)
print(df_retiros)

print("\nDatos de la tabla 'general':")
df_general = pd.read_sql("SELECT * FROM general", conn)
print(df_general)

print("\nDatos de la tabla 'employee':")
df_employee = pd.read_sql("SELECT * FROM employee", conn)
print(df_employee)

print("\nDatos de la tabla 'manager':")
df_manager = pd.read_sql("SELECT * FROM manager", conn)
print(df_manager)

# Realizar consultas SQL para análisis
print("\nConteo de 'Attrition' en la tabla 'retiros':")
attrition_count = pd.read_sql("SELECT Attrition, COUNT(*) FROM retiros GROUP BY Attrition", conn)
print(attrition_count)

print("\nConteo de 'BusinessTravel' en la tabla 'general':")
business_travel_count = pd.read_sql("SELECT BusinessTravel, COUNT(*) FROM general GROUP BY BusinessTravel", conn)
print(business_travel_count)

print("\nConteo de 'JobSatisfaction' en la tabla 'employee':")
job_satisfaction_count = pd.read_sql("SELECT JobSatisfaction, COUNT(*) FROM employee GROUP BY JobSatisfaction", conn)
print(job_satisfaction_count)

```

## ELIMINACION DE NULOS

```python
# Revisar columnas con valores nulos en df_mergedtriple
valores_nulos = df_mergedtriple.isnull().sum()

# Filtrar y mostrar solo las columnas que tienen valores nulos
columnas_con_nulos = valores_nulos[valores_nulos > 0]

# Mostrar las columnas con valores nulos y el número de valores nulos en cada una
print(columnas_con_nulos)
```

```python
# Lista de columnas a las que se les reemplazarán los nulos por la mediana
columns_to_replace = ['NumCompaniesWorked', 'TotalWorkingYears', 'EnvironmentSatisfaction', 'JobSatisfaction', 'WorkLifeBalance']

# Reemplazar valores nulos con la mediana de cada columna
for column in columns_to_replace:
    median_value = df_mergedtriple[column].median()
    df_mergedtriple[column].fillna(median_value, inplace=True)

# Verificar que no haya valores nulos en las columnas seleccionadas
print(df_mergedtriple[columns_to_replace].isnull().sum())
```

```python
# Eliminar las columnas 'retirementType' y 'resignationReason' de df_mergedtriple. se eliminan porque estas variables puedes sergar la solucion del modelo, ya que estan relcionadas con la variable explicativa que es atrition
df_mergedtriple.drop(columns=['retirementType', 'resignationReason'], inplace=True)

# Verificar que las columnas se hayan eliminado
print(df_mergedtriple.head())
```

```python
# Eliminar las columnas 'index_x', 'index_y' y 'index_xindex_y' de df_mergedtriple
df_mergedtriple = df_mergedtriple.drop(columns=['index_x', 'index_y','index'], errors='ignore')

# Mostrar las primeras filas del DataFrame limpio para verificar
print(df_mergedtriple.head())
df_mergedtriple
```

```python
#attrition es la variable objetivo asi que se le da este tratamiento para posteriormente incluirla en el modelo

# Reemplazar 'Yes' por 1 en la columna 'Attrition'
df_mergedtriple['Attrition'] = df_mergedtriple['Attrition'].replace({'Yes': 1})

# Llenar los valores nulos en la columna 'Attrition' con 0
df_mergedtriple['Attrition'] = df_mergedtriple['Attrition'].fillna(0)

df_mergedtriple['Attrition'] = df_mergedtriple['Attrition'].astype(int)

# Mostrar las primeras filas del DataFrame para verificar
print(df_mergedtriple.head())
df_mergedtriple
```

## ELIMINACION DE VARIABLES RESTANTES


```python
# Suponiendo que df_mergedtriple ya está cargado y listo para usar

# Lista de variables a analizar
variables = ['EmployeeCount', 'Over18', 'StandardHours', 'JobRole', 'Department', 'Gender', 'PerformanceRating']

# Análisis para cada variable
for var in variables:
    print(f"Análisis de la variable: {var}")

    # Información de la variable
    print(f"\nInformación sobre '{var}':")
    print(df_mergedtriple[var].info())

    # Resumen de la variable
    print(f"\nResumen de '{var}':")
    if pd.api.types.is_numeric_dtype(df_mergedtriple[var]):
        print(df_mergedtriple[var].describe())
    else:
        print(df_mergedtriple[var].value_counts(dropna=False))

    print("\n" + "="*50 + "\n")
```

```python
# Razones para eliminar cada columna:
# 'EmployeeCount': Esta columna tiene un valor constante de 1 para todos los registros y no aporta información útil.
# 'Over18': Esta columna tiene un valor constante de 'Y' para todos los registros, lo que no proporciona variabilidad.
# 'StandardHours': Esta columna tiene un valor constante de 8 para todos los registros y no varía entre empleados.
# 'JobRole': Aunque es una variable relevante, la eliminaríamos si se desea simplificar el DataFrame ya que contiene 9 categorias diferentes.
# 'Department': Similar a 'JobRole', la eliminaremos si se busca simplificar el DataFrame.
# 'Gender': Aunque es relevante, se eliminaría en caso de simplificar el DataFrame.
# 'PerformanceRating': Aunque importante para el análisis, se eliminaría ya que la simplificación del DataFrame es primordial .

# Eliminar las columnas
df_mergedtriple = df_mergedtriple.drop(columns=['EmployeeCount', 'Over18', 'StandardHours', 'JobRole', 'Department', 'Gender', 'PerformanceRating'])

# Mostrar el DataFrame resultante para verificar los cambios
print(df_mergedtriple.head())
df_mergedtriple
```
